```python
import pandas as pd
```

```python
df=pd.DataFrame({"Name":['jay','sam','roy','amar','sunil','aman','nitin','roshan','chunil','chotu'],
                 "Salary":[40000,20000,50000,10000,30000,35000,50000,80000,12000,82000],
                 "Country":["India","Bangladesh","India","India","bhutan","England","Italy","USA","Canada","Ukraine"]})
```

```python
df
```

```python
print("Mean :",df.Salary.mean())
print("Median :",df.Salary.median())
print("Mode :",df.Salary.mode()[0])
```

```python
print("Variance :", df.Salary.var())
print(" Standard Deviation :", df.Salary.std())
```

```python
print("Maxmium Salary :",df.Salary.max())
print("Minimum Salary :",df.Salary.min())
```

```python
df.groupby("Country")["Salary"].sum().reset_index(name="Total_Salary")
```

```python
print("Skewness :", df.Salary.skew())
print("Kurtosis :", df.Salary.kurtosis())
```
