```python
plt.figure(figsize=[15,10])
totalDeath=[113055,37312,59751,75473,33964]
totalRecovery=[430480,325602,230688,129095,166584]
activeCases=[139958,347973,23999,12960,34730]
country=['USA','Brazil','Russia','India','Italy']
x=np.arange(len(totalDeath))
plt.bar(x,totalDeath,color='black',width=0.25)
plt.bar(x+0.25,totalRecovery,color='g',width=0.25)
plt.bar(x+0.5,activeCases,color='b',width=0.25)
plt.legend(['total Deaths','total Recovery','Active cases'])
plt.show()
```

```python
#amans code
countries=["India", "USA", "Brazil", "Russia", "Spain", "UK"]
total_cases=[20000, 40000, 8000, 50000, 4000, 10000]
total_deaths=[1880, 37312, 5971, 45789, 3488, 9008]
 
X_axis = np.arange(len(countries))
 
plt.bar(X_axis - 0.2, total_cases, 0.4, label = 'Total cases',color="red")
plt.bar(X_axis + 0.2, total_deaths, 0.4, label = 'Total deaths',color="grey")
 
plt.xticks(X_axis, countries)
plt.xlabel("Countries")
plt.ylabel("Number of people")
plt.title("Deaths and cases")
plt.legend()
plt.show()
```
